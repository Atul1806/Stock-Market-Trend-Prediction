```python
#Splitting data into training and testing
data_train=pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_test=pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])
print(data_train.shape)
print(data_test.shape)
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
data_train_array=scaler.fit_transform(data_train)
data_train_array.shape
```

```python
d=[]
for i in data_train_array:
    d.append(float(i))
print(d)
```

```python
data_train
```

```python
len(d)
```

```python
x_train=[]
y_train=[]
for i in range(100,data_train_array.shape[0]):
    x_train.append(d[i-100:i])
    y_train.append(d[i])
x=x_train
x_train,y_train=np.array(x_train),np.array(y_train)
```

```python
x_train.shape
```

```python
y_train
```

```python
x_train.shape
```

```python
from sklearn.tree import DecisionTreeRegressor 
  
# create a regressor object
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x_train,y_train)
```

```python
data_test.head()
```

```python
past_100=data_train.tail(100)
past_100
```

```python
data_test.shape
```

```python
final_df=past_100.append(data_test,ignore_index=True)
```

```python
final_df
```

```python
input_data=scaler.fit_transform(final_df)
```

```python
input_data.shape
```

```python
data_test_array=scaler.fit_transform(input_data)
data_test_array.shape
```

```python
h=[]
for i in data_test_array:
    h.append(float(i))
len(h)
```

```python
x_test=[]
y_test=[]
for i in range(100,input_data.shape[0]):
    x_test.append(h[i-100:i])
    y_test.append(h[i])
    
```

```python
x_test,y_test=np.array(x_test),np.array(y_test)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
y_test.shape
```

```python
#Making Predictions

y_pred = regressor.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
scaler.scale_ #Factor by which all these values are scaled down
```

```python
scale_factor=1/0.00690691
y_predict=y_pred*scale_factor
y_test=y_test*scale_factor
```

```python
y_test
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test,'b',label='Original Price')
plt.plot(y_predict,'r',label='Predicted Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2)) 
print("R2 score =", round(sm.r2_score(y_test, y_pred), 2))
```
